# Matplotlib

caracteristicas:
<ul>
    <li>-es un modulo de python que permite la creacion de visualizacion 2d y 3d basada en datos</li>
    <li>-matplotlib no define una gramatica de graficos en la que se formalicen los componentes de un grafico basado en datos</li>
    <li>-ofrece un conjunto amplio de funciones (atraves del submodulo mtplotlib.plt) que permite crear diferentes tipos de graficos (puntos, barras, lineas, etc )</li>
    <li>-la llamada a estas funciones no genera un grafico, sino que modifican el estado de n objeto interno y comun a todas las llamadas</li>
    <li>-la utilizacion de la funcion show ofrecida por el modulo sera la que se encargue de renderizar el objeto interno y comun en el estado que se encuentre</li>
    <li>-permite, por tanto, la creacion iterativa de graficos</li>
</ul>

# Nuestro primer grafico

creemos nuestro primer grafico. Vamos a crear un grafico de linea representando la funcion f(x) = x**2. Podemos hacerlo mediante la fincion plot, la mas basica ofrecida por matplotlib

Ahora, llamamos a la función`.plot` la cual va a recibir los siguientes argumentos:


*   `x`: Eje X. Opcional
*   `y`: Eje y
*   `fmt`: Formato del gráfico en cuanto a los puntos, líneas y colores ('[marker][line][color]') . Opcional
*   Inputs de formato de las líneas opcionales. Los más usuales:
  * `label`: Para nombrar cada línea
  * `color`: Color en formato numérico si fuera necesario
  * `linewidth`: Ancho de línea
  * `markersize`: Tamaño de los puntos


```python
x_values = range(10)
y_values = [value ** 2 for value in x_values]
plt.plot(x_values, y_values)
```

# Entendiendo la salida de la funcion plot

Podemos ver que  la salida contiene dos elementos
<li>Primero un objeto grafico de tipo Line2D representado por la cadena entre corchetes</li>
<li>Segundo la representacion grafica de dicho objeto</li>

Esta salida doble se da por el modod de fucnionamiento interactivo establecido actualmente por matploit. El modo interactivo hace que el paquete lleve a cabo el renderizado de forma automatica (sin llamar a la funcion show).
Sin embargo el modo interacitco de matplotlib es configurable. Si no se especifica nada, por defecto se establecera en funcion del tipo de entorno que estemos utilizando
<li>Interactivo: en notebook, idees graficos como Rodeo o Spyder</li>
<li>No Interactivo: En consolas de comandos como python o ipython y en ejecucion de scripts .py completos</li>

Para comprobar comprobar el modo de funcionamiento establecido y modificarlo, matplotlib nos da tres funciones

```python
#comprobar el modo de funcionamiento de matplotlib
print(plt.isinteractive())
```

```python
#Desactivar el modo interactivo
plt.ioff()
print(plt.isinteractive())
```

```python
#Activar el modo interactivo
plt.ion()
print(plt.isinteractive())
```

# Un grafico con mas componentes
Para comprender cual es la filosofia de trabajo con matplotlib vamos a tratar de crar un grafico un poco mas completo. Este grafico tendra:
<ul>
<li>Una serie que contenga la funcion seno</li>
<li>Una serie que contenga la funcion coseno</li>
<li>Nombres en ambos ejes</li>
<li>Una leyenda que indique que linea representa cada funcion</li>
</ul>
Para facilitar la creacion deñ set de datos vamos a utilizar numpy

```python
import numpy as np
#Creamos nu8estro set de datos
x_values = np.linspace(0, 2 * np.pi, 100)
sin_y_values = np.sin(x_values)
cos_y_values = np.cos(x_values)

#añadimos la serie de la funcion seno 
sin_line, = plt.plot(x_values, sin_y_values)
cos_line, = plt.plot(x_values, cos_y_values)
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend([sin_line, cos_line], ['y = sin(x)', 'y = cos(x)'])
plt.show()
sin_line, = plt.plot(x_values, sin_y_values)
plt.show()
```

# Tipos de graficos

En este notebook vamos a revisar algunas de las funciones que nos ofrece matplotlib para la generacion de diferentes tipos de graficos. Como se ha comentado, estas funciones siguen el estilo de basi-r teniendo cada una de ellas un conjunto amplioo de parametros que permiten definir todas sis caracteristicas (desde datos, a colores, estilos, etc)

Como se ha visto en ejemplos anteriores, estas funciones se pueden utilizar de forma conjunta como diferentes componentes o capas de un determinado grafico. Esta composicion se realiza mediante la llamada seciencial "en orden" y se podra obtener el resultado de la composicion mediante la llamada a la funcion show.


Como no tengo el archivo con esa info esta parte no la pude hacer siguiendo el curso pero voy a poner cada uno de los distintos graficos sin ejemplificar
<ul>
<li>Primero que nada podemos modificar las lineas y su puntos como queramos, linea punteada, linea no punteada etc. Tambien los colores de la lineas o rellenos o puntos.</li>
<li><b>Area:</b> en este caso nos encontramos de nuevo con graficos que en este caso el area por debajo o encima si los valores son negativos, de la linea del eje X se ira relleno. La funcion ofrecida para este tipo de grafico se llama stackplot y sus parametros son identicos a los de plot. <b>plt.stackplot(parametro1,parametro2)</b></li>
<li><b>Puntos(scatter):</b> este es uno de los graficos mas utilizados. Matplotlib nos ofrece la funcion scatter. Esta funcion recibira como minimo un conjunto de valores para el eje de las X y un conjunto de valores para el eje de las Y. Adicionalmente nos ofrece la posibilidad de controlar diferentes caracteristicas visuales de los puntos represetados: tamaño, alpha, color, tipo de punto, etc. <b>plt.scatter(parametro1, parametro2, color='black', alpha=0)</b></li>
<li><b>Barras:</b> para la creacion de graficos de barras matplotlib nos ofrece dos funciones bar y barh (en funcion e la orientacion que queremos dar al grafico). Para su creacion tambien tenemos que usar un conjunto de valores para el eje X y el conjunto de valores para el eje Y. Tambien tenemos diferentes caracteristicas para modificar las barras visualmente. <b>plt.bar(parametro1,parametro2)</b></li>
<li><b>tarta:</b> en este caso nos ofrece la funcion pie, para la construccion de este grafico sera necesario suministrar a la funcion un conjunto unico de valores de forma que el modulo presentara la proporcion de cada uno de los vlaores suministrados sobre el total. Tmabien nos permite modificar caracteristicas visuales. <b>plt.pie(datset)</b></li>
<li><b>Histogramas:</b> mediante la funcion hist, matplotlib nos permite llevar a cabo la creacion de histogramas para la representacion de la distribucion de una variable numerica. Este conjunto de valores numericos sera el unico parametro obligatorio necesario para la creacion del grafico. Adicionalmente dispondremos de parametros para controlar: numero de bins del histograma, normalizacion del histograma (de froma que obtengamos densidades), indicacion de si se debe realizar un histograma acumulativo, etc. ej: <b>plt.hist(diamantes['price'])</b></li>
<li><b>Box-plot:</b> disponemos de la funcion boxplot que recibira un conjunto de valores sobre el que calcular los rangos, medianas, bigotes y outliers. Dispondremos de un conjunto de parametros que nos permite controlar: tipo de punto para representar outliers, control de si se muestran o no cajas y bigotes, etc. ej:<b>plt.boxplot(diamons['price'])</b></li>
<li><b>Violin:</b> los graficos violin son un hibrido entre boxplot y curvas de densidad.Para la construccion de ests tipo de graficos, dispondremos de la funcion violinplot a la que habra que suministrar siempre el conjunto de valores sobre el que se quiere calcular la distribucion. Dispondremos de un conjunto de parametros para modificar sus caracteristicas: orientacin, numero de puntos utilizadi para el calculo de la distribucion, metodo de calculo, etc. ej: <b>plt.violinsplot(diamons['price'])</b></li>
<li><b>Rango de linea:</b> para llevar a cabo la construccion de este tipo de graficos, matplotlib pone a nuestra disposicion la funcion vlines, a la que habra que suministrar por tanto 3 series de datos (uno para cada coordenada de la linea). Ej: <b>plt.vinlines(min_max_x.index, min_max_x['min'], min_max_x['max'])</b></li>
